# Data Standardization

The process of standardizing the data to a common format and common range.

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Loading the dataset
dataset_bc = sklearn.datasets.load_breast_cancer()
```

```python
# Loading the dataset to a DataFrame
df_bc = pd.DataFrame(dataset_bc.data, columns = dataset_bc.feature_names)
```

```python
df_bc.head()
```

```python
df_bc.shape
```

```python
# Storing the DataFrame and target values in different variables.
X = df_bc
Y = dataset_bc.target
```

# Splitting The Data Into Training Data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Standardizing The Data

```python
# If the standard deviation of dataset is 1 or closer to 1 then all the values of the dataset is in same range.
print(dataset_bc.data.std())
```

```python
scaler = StandardScaler()
# Analyzing how data is distributed in X_train
scaler.fit(X_train)
```

```python
# Transforming the X_train data based on the standard scaler
X_train_stndrd = scaler.transform(X_train)
```

```python
print(X_train_stndrd)
```

```python
# Transforming the X_test data based on X_train Standardization
X_test_stndrd = scaler.transform(X_test)
```

```python
print(X_test_stndrd)
```

```python
# Rechecking the standard deviation of X_train
print(X_train_stndrd.std())
```

```python
# Standard deviation of X_test
print(X_test_stndrd.std())
# We get 0.865... because we are transforming on the basis of X_train, but it is way better than 228.29...
```
